```python
import re
import urllib
from bs4 import BeautifulSoup
def PageCrawler(recipeUrl):
    url = "http://www.10000recipe.com/" + recipeUrl
    req = urllib.request.Request(url)
    sourcecode = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(sourcecode, "html.parser")
    
    error_return = [{"\[재료\]":[]}]
    
    recipe_title = []
    recipe_source = {}
    recipe_tag = {}
    ingredient = {}
    res = soup.find('div','view2_summary')
    if res == None:
        return error_return
    else:
        res = res.find('h3')
    recipe_title.append(res.get_text())
    res = soup.find('div','view2_summary_info')
    recipe_title.append(res.get_text().replace('\n',''))
    
    res = soup.find('div','ready_ingre3')
    
    try:
        for n in res.find_all('ul'):
            for tmp in n.find_all('li'):
                ing = tmp.get_text().split('\n')[0].replace(' ','')
                if len(tmp.find_all('span')) == 0:
                    ingredient[ing] = ""
                else:
                    for span in tmp.find_all('span'):
                        sp = span.get_text().split('\n')[0].replace(' ','').lstrip().rstrip()
                        if ing in ingredient:
                            if len(re.findall("\d+", ingredient[ing])) == 1 and len(re.findall("\d+", sp)) == 1:
                                ingredient[ing] = str(int(re.findall("\d+", ingredient[ing])[0]) + int(re.findall("\d+", sp)[0])) + re.sub(r'[0-9]+', '', sp)
                            elif len(re.findall("\d+", ingredient[ing])) > 0:
                                ingredient[ing] = ""
                        else:
                            ingredient[ing] = sp

    except (AttributeError):
        return
    '''
    if(res.find('div','view2_tag')):
        recipe_tag = res.find('div','view2_tag').get_text()
        print(recipe_tag)
        recipe_all = [recipe_title, recipe_source,recipe_tag,{'id':recipeUrl}]
    '''
    #recipe_all = [recipe_title, recipe_source,{'id':recipeUrl}]
    recipe_all = {}
    recipe_all[recipeUrl.split('/')[1]] = ingredient
    print(f"{recipeUrl} scraping complete!")
    return(recipe_all)
```

```python
import re
regex = "\(.*\)|\s-\s.*"
def matching(x):
    matchlist = ["고등어","고구마","고들빼기","고사리","고수","고추가루","고추장아","고형카레","강황","기름","간장게장","가쓰오부시","갈비양념","계란",
                '꽃게',"꽈리고추","느타리버섯","닭고기","당근","대구","동치미","돼지고기","들기름","떡볶이떡","라면","로즈","대파"]
    returnNothing = ["기용","기태","기호","길이","금기","기타"]
    returnNothingSearch= ["컵","숟가락","재료","생략가능","선택사항","나무","내장","냉장고","꼬치"]
    returnSearch = ["김치","까나리","깍두기","꿀","깨소금"]
    for i in matchlist:
        if re.match(i,x):
            return i
        
    for i in returnNothing:
        if re.match(i,x):
            return ""
        
    for i in returnNothingSearch:
        if re.search(i,x):
            return ""
    for i in returnSearch:
        if re.search(i,x):
            return i
        
    if re.match("건조",x):
        return x[2:]
    elif re.search("컵",x) or re.search("숟가락",x):
        return ""
    elif re.match("국산",x) or re.match("국수",x) or re.match("데친",x):
        return x[2:]
    #elif re.match("굴소스",x):
    #    return x[3:]
    elif re.match("고기",x):
        if re.match("고기육수",x):
            return "고기육수"
        return "고기"
    elif re.search("불고기",x):
        if re.search("고추장",x):
            return x
        return "불고기"
    elif re.match("다시마",x):
        if re.search("다시마미역",x):
            return "다시마미역"
        elif re.search("다시마멸치",x):
            return "다시마멸치"
        return "다시마"
    elif re.match("닭가슴",x):
        return "닭가슴살"
    elif re.match("닭볶음",x):
        return "닭볶음닭"
    elif re.match("당면",x):
        if re.search("당면사리",x):
            return "당면사리"
        elif re.search("당면순대",x):
            return "당면순대"
        return "당면"
    elif re.match("돈가스",x):
        return "돈까스"
    elif re.match("레몬",x):
        if re.search("레몬주스",x):
            return x
        return "레몬"
    else:
        return x

    
def prematch(item):
    sp = re.split(r'[0-9]',str(item))
    sp[0] = re.sub(regex, '', sp[0])
    sp[0] = sp[0].replace("조금","").replace("약간","").replace("적당히","").replace("적당량","").replace("구운","").replace("원하는만큼","").replace("많이","").replace("다진","").replace("먹을만큼","").replace("(","").replace(")","")
    sp[0] = sp[0].replace("큰것","").replace("작은것","").replace("살짝","").replace("취향껏","").replace("취향것","").replace("▶재료:","").replace("아주","").replace("솔솔","").replace("톡톡","").replace("기호에맞게","").replace("취향에맞게","")
    sp[0] = sp[0].replace("작은","").replace("큰","").replace("소량","").replace("대량","").replace("크게","").replace("껍질벗긴","").replace("손질한","").replace("말린","").replace("남은것","").replace("얼린","").replace("입맛대로","").replace("마른","")
    sp[0] = re.split('또는',sp[0])[0]
    sp[0] = re.split('or',sp[0])[0]
    sp[0] = re.sub(r'\u200b',r'',sp[0])
    sp[0] = re.sub(r'\u202d',r'',sp[0])
    sp[0] = re.split('[^가-힣]',sp[0])[0]
    sp[0] = sp[0].replace("고춧","고추").replace("큰술","").replace("작은술","").replace("적당량","").replace("약간","").replace("조금","").replace("양파거","양파")
    sp[0] = sp[0].replace("달걀","계란")
    sp[0] = sp[0].replace("쭈꾸미","주꾸미")
    sp[0] = sp[0].replace("찬","").replace("뜨거운","").replace("더운","").replace("다진","").replace("불린","")
    sp[0] = sp[0].replace("가나","").replace("는둥만둥","").replace("각각","").replace("가는","").replace("자른","").replace("가다랭어포","가다랭이포").replace("가다랭이포","가다랑이포")
    sp[0] = sp[0].replace("가스오","가쓰오").replace("적당","").replace("국물용말고얇은거","").replace("가위","").replace("것","").replace("썬거","").replace("안찐","").replace("안찐것","").replace("찐거","")
    sp[0] = matching(sp[0])
    if re.search("만큼",str(sp[0])) or re.search("접시",str(sp[0])) or re.search("이쑤시개", str(sp[0])) or str(sp[0]) == "":
        return "hastoerase"
    else:
        if sp[0] == 'LA갈비':
            return sp[0]
        else:
            sp[0] = re.split('[a-z]',str(sp[0]))[0]
            return sp[0]

prematch("현미유기름")
```
